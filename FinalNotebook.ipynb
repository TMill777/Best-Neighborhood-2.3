## Introduction

When we were tasked with narrowing down the best neighborhood in Pittsburgh, we used the following WPRDC datasets:

- Fire Incidents
- Arrest Report
- County Assets

We balanced a priority of safety with the fire incidents and the arrest report, with the last data set of county assets showing that the best neighborhood will have a combination of safety + assets. Fire incidents was a large indicator towards safety in a surrounding area because if theres a high number of fire incidents, it could indicate a couple of things, such as a lackluster fire department or irresponsible people, both of which could negatively impact a neighborhood a lot. We also have the data for arrests in each neighborhood, which is obviously a huge indicator towards neighborhood safety. The safest neighborhoods will have a low arrest number combined with a low fire incident count. However, we didn't just measure safety, and also measured the amount of assets (businesses/schools/healthcare services/etc) that each neighborhood has. A neighborhood could be very safe but doesn't have the necessary assets that truly make it the best neighborhood, so the true best neighborhood will be proficient in all three categories.
