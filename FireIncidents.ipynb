```python
#importing data and showing first few rows
import pandas as pd

url = "https://data.wprdc.org/datastore/dump/8d76ac6b-5ae8-4428-82a4-043130d17b02"
fire_incidents_df = pd.read_csv(url)
fire_incidents_df.head()
```

```python
#gets numeral statistics
fire_incidents_df.describe()
```

```python
#this code steps through the data to find the 10 neighborhoods that have the fewest fire incidents
neighborhood_incidents = fire_incidents_df['neighborhood'].value_counts()

print("10 Neighborhoods With Least Fire Incidents:\n")
print(neighborhood_incidents.tail(10))

safest = neighborhood_incidents.idxmin()
print(f"\nThe Safest Neighboorhood for Fire Incidents is: {safest}")
```

```python
#this code goes through the data again to get the incidents for ALL neighborhoods and displays from most to least
neighborhood_incidents = fire_incidents_df['neighborhood'].value_counts()

print("Neighborhoods Ranked By Number of Total Fire Incidents (Most to Least):\n")
for neighborhood, count in neighborhood_incidents.items():
    print(f"{neighborhood}: {count} incidents")
```

```python
#this code simply shows the top 3 fire incident types in pittsburgh through a matplot graph (113 is cooking fire, 11 is whole building fire, and 131 is vehicle fire)
import matplotlib.pyplot as plt

top_incidents = fire_incidents_df['incident_type'].value_counts().head(3)
top_incidents.plot(kind='bar', title='Top 3 Fire Incident Types in Pittsburgh', figsize=(10,3))
plt.xlabel('Incident Type')
plt.xticks([0, 1, 2], ['Cooking Fire', 'Building Fire', 'Vehicle Fire'])
plt.ylabel('Count')
plt.show()
```
